```python
# cofre digital

# O cofre possui um codigo composto por 5 números inteiros

# desenvolva um programa que:

# leia os 5 números utilizando laços de repetição

# ao final informar:

# 1 - Quais números são pares
# 2 - Quais números são imparees
# 3 - O maior número
# 4 - A soma de todos os números
# 5 - A média dos números
# 6 - O menor número
# 7 - Quantos números são maiores que a média

pares = 0
impares = 0
soma = 0

for i in range(5):
    num = int(input("Digite um número inteiro para o cofre: "))

    if i == 0:
        n1 = num
    elif i == 1:
        n2 = num
    elif i == 2:
        n3 = num
    elif i == 3:
        n4 = num
    elif i == 4:
        n5 = num

    soma = soma + num

    if num % 2 == 0:
        pares = pares + 1
    else:
        impares = impares + 1

media = soma / 5

maior = n1
if n2 > maior: maior = n2
if n3 > maior: maior = n3
if n4 > maior: maior = n4
if n5 > maior: maior = n5

menor = n1
if n2 < menor: menor = n2
if n3 < menor: menor = n3
if n4 < menor: menor = n4
if n5 < menor: menor = n5

maiores_que_a_media = 0
if n1 > media: maiores_que_a_media = maiores_que_a_media + 1
if n2 > media: maiores_que_a_media = maiores_que_a_media + 1
if n3 > media: maiores_que_a_media = maiores_que_a_media + 1
if n4 > media: maiores_que_a_media = maiores_que_a_media + 1
if n5 > media: maiores_que_a_media = maiores_que_a_media + 1

print()
print("--- RELATÓRIO DO COFRE ---")
print("1 - Quantos números são pares:", pares)
print("2 - Quantos números são ímpares:", impares)
print("3 - O maior número:", maior)
print("4 - A soma de todos os números:", soma)
print("5 - A média dos números:", media)
print("6 - O menor número:", menor)
print("7 - Quantos números são maiores que a média:", maiores_que_a_media)
```

```python
# Número espelho

# Solicite um número inteiro positivo

# Inverta o número
# informe se ele é um palíndromo.

numero_texto = input("Digite um número inteiro positivo: ")

numero_invertido = numero_texto[::-1]

print("Número invertido:", numero_invertido)

if numero_texto == numero_invertido:
    print("O número é um palíndromo!")
else:
    print("O número não é um palíndromo.")
```
