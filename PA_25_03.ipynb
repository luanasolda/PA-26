```python
#escrever um programa que recebe 04 notas do aluno, calcula a media e mostra se o aluno foi aprovado

#entrada
print("digite suas notas AQUI, valores 01 a 10")
nota1 = float(input("digite a nota do 1º bimestre: "))
nota2 = float(input("digite a nota do 2º bimestre: "))
nota3 = float(input("digite a nota do 3º bimestre: "))
nota4 = float(input("digite a nota do 4º bimestre: "))

#processamento
resultado = (nota1 + nota2 + nota3 + nota4) / 4
if resultado >= 7:
  print(f"A mèdia das suas notas foi {resultado}, você foi aprovado")
elif resultado >= 5:
  print(f"A média das suas notas foi {resultado}, você terá que fazer recuperação")
else:
  print(f"A média das suas notas foi {resultado}, você foi reprovado")
```

```python
#Fazer uma calculadora simples com as 04 operações
print("------MENU-----")
print("")
print(" opção - 1 = SOMA")
print(" opção - 2 = SUBTRAÇÃO")
print(" opção - 3 = MULTIPLICAÇÃO")
print(" opção - 4 = DIVISÃO")
print("")

opicao = int(input("digite a opção: "))

#entrada
num1 = int(input("digite o primeiro numero: "))
num2 = int(input("digite o segundo numero: "))

#processamento

if opicao == 1:
  resultado = num1 + num2
  print(f"A soma de {num1} mais {num2} é igual a {resultado}")
elif opicao == 2:
  resultado = num1 - num2
  print(f"{num1} menos {num2} é igual a {resultado}")
elif opicao == 3:
  resultado = num1 * num2
  print(f"A multiplicação de {num1} por {num2} é igual a {resultado}")
elif opicao == 4:
    resultado = num1 / num2
    print(f"A divisão de {num1} por {num2} é igual a {resultado}")
```
