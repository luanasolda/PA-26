```python
#PAR OU IMPAR

#entrada
num = int(input("Digite um número: "))

#processamneto
if num % 2 == 0:
  print(f"{num} é um número par.")

else:
  print(f"{num} é um número impar.")
```

```python
#MAIOR DE DOIS NÚMEROS
#entrada
a = int(input("Digite o primeiro número: "))
b = int(input("Digite o segundo número: "))

#processamento
if a > b:
  print(f"o número {a} é maior que o número {b}")

elif b > a:
  print(f"o número {b} é maior que o número {a}")

else:
  print(f"os dois números são iguais")
```

```python
#APROVAÇÃO DE ALUNO
#entrada
nota1 = float(input("Digite sua nota do 1° bimestre: "))
nota2 = float(input("Digite sua nota do 2° bimestre: "))
nota3 = float(input("Digite sua nota do 3° bimestre: "))
nota4 = float(input("Digite sua nota do 4° bimestre: "))

#processamento
media = (nota1 + nota2 + nota3 + nota4) / 4

#saida
if media >= 7:
  print(f" Sua média éde {media}, e você foi aprovado.")
elif media >= 5:
  print(f"Sua média é de {media}, e você ficou de recuperação.")
elif media < 5:
  print(f"Sua média é de {media}, e você foi reprovado.")
```

```python
#CLASSIFICAÇÃO DE IDADE
print("Criança (0–12)")
print("Adolescente (13–17)")
print("Adulto (18–59)")
print("Idoso (60+)")
print("")

#entrada
a = int(input("Digite sua idade: "))

#processamento
if a <= 12:
  print(f"Você tem {a} anos, e ainda é uma criança.")
elif a > 12 and a <= 17:
  print(f"Você tem {a} anos, e ja é um adolecente.")
elif a >= 18 and a < 60:
  print(f"Você tem {a} anos, e é um adulto")
else:
  print(f"Você tem {a} anos, e é um idoso")
```

```python
#POSITIVO, NEGATIVO OU ZERO
#entrada
a = int(input("Digite um número: "))

#processamento
if a < 0:
  print(f"{a} é um número negativo.")
elif a > 0:
  print(f"{a} é um número positivo.")
elif a == 0:
  print(f"{a} é igual a zero")
```

```python
#CALCULADORA SIMPLES
print("---operações---")
print("1 = adição")
print("2 = subtração")
print("3 = multiplicação")
print("4 = divisão")
print("")

#entrada
op = int(input("Digite uma operação: "))
print("")
a = int(input("Digite o primero número: "))
b = int(input("Digite o segundo número: "))
#processamento
if op == 1:
  res = a + b
  print(f"o resultado de {a} mais {b} é igual a {res}")
elif op == 2:
  res = a - b
  print(f"o resultado de {a} menos {b} é igual a {res}")
elif op == 3:
  res = a * b
  print(f"O resultado de {a} vezes {b} é igual a {res}")
elif op == 4:
  res = a / b
  print(f"O resultado de {a} dividido por{b} é igual a {res}")
```

```python
#DIA DA SEMANA
#entrada
d = int(input("Digite um número: "))

#processamento
if d == 1:
  print(f"{d} é igual a domigo")
elif d == 2:
  print(f"{d} é igual a segunda")
elif d == 3:
  print(f"{d} é igual a terça")
elif d == 4:
  print(f"{d} é igual a quarta")
elif d == 5:
  print(f"{d} é igual a quinta")
elif d == 6:
  print(f"{d} é igual a sexta")
elif d == 7:
  print(f"{d} é igual a sabado")
```

```python
#ANO BISSEXTO

#entrada
a = int(input("Digite um ano: "))

#processamento
if a % 4 or a % 400:
  print(f"O ano {a} é um ano bissexto")
else:
  print(f" O ano {a} não é bissexto")
```
