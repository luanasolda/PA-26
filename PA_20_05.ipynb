```python
# você foi contratado para desenvolver um sistema simples de caixa eletrônico
# o sistema deverá funcionar atravez de um menú interativo utilizando o sistema de reptição do while
# o programa deve iniciar com um saldo de R$1.000,00 e permitir que o usuario escolha diferentes operações até desistir sair do sistema

# funcionalidades:
# ver saldo
# depositar dinheiro
# sacar dinheiro
# sair

#Regras:
# ver saldo
# Ao escolher essa opição o sistema deverá mostrar o saldo atual da conta

# Depositar dinheiro
# o usuario deverá informar um valor para deposito
# o valor digitado deverá ser somado ao saldo da conta

# sacar dinheiro
# o usuario deverá imformar o valor de saque

# O sistema deverá verificar:
# se houver suficiente - realizar saque
# se não houver saldo suficiente - mostrar mensagem de "saldo insuficiente"

# sair
# o sistema deverá encerrar apenas quando oisiario escolher a opção

saldo = 1000.00
opcao = 0

while opcao != 4:
    print("\n--- CAIXA ELETRÔNICO ---")
    print("1 - Ver saldo")
    print("2 - Depositar dinheiro")
    print("3 - Sacar dinheiro")
    print("4 - Sair")

    opcao = int(input("\nEscolha uma opção: "))

    if opcao == 1:
        print(f"Seu saldo atual é: R$ {saldo:.2f}")

    elif opcao == 2:
        valor_deposito = float(input("Digite o valor do depósito: R$ "))
        if valor_deposito > 0:
            saldo += valor_deposito
            print("Depósito realizado com sucesso!")
        else:
            print("Valor inválido para depósito.")

    elif opcao == 3:
        valor_saque = float(input("Digite o valor do saque: R$ "))
        if 0 < valor_saque <= saldo:
            saldo -= valor_saque
            print("Saque realizado com sucesso!")
        else:
            print("Saldo insuficiente ou valor inválido.")

    elif opcao == 4:
        print("Encerrando o sistema. Até logo!")

    else:
        print("Opção inválida. Tente novamente.")
```
