#Escrever um programa que faz a conversão de temperatura

entrada: Temperatura em Celsius
Processamento: converter para Fahrenheit
saida: resultado
Formula:F = (C x 9/5) + 32

```python
#Entrada

temperatura = float(input("digite a temperatura: "))

#Processamento

Fahrenheit = (temperatura * 9/5) + 32

#saida

print (f"a temperatura e {Fahrenheit}")
```
