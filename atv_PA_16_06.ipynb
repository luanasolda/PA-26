```python
#QUESTÃO 01 – Contagem Crescente (2 pontos)

#Elabore um programa que exiba na tela os números de 1 até 20,
#um em cada linha.

#Exemplo de saída:
1
2
3
...
20


n = int(input("Digite um numero: "))

#processamento
for i in range(n):
  print(i + 1)
```

```python
#QUESTÃO 02 – Soma dos Números (2 pontos)

#Solicite ao usuário um número inteiro positivo.

#Utilizando um laço de repetição, calcule e exiba a soma de
#todos os números de 1 até o número informado.

#Exemplo:

#Digite um número: 5

#Resultado: 15

#Pois:

#1 + 2 + 3 + 4 + 5 = 15

#entrada

n = int(input("Digite um número inteiro positivo: "))

soma = 0

for i in range(1, n + 1):
  soma += i

print(f"A soma dos números de 1 até {n} é: {soma}")
```

```python
#QUESTÃO 03 – Tabuada (2 pontos)

#Solicite ao usuário um número inteiro e exiba sua tabuada
#de 1 até 10.

#Exemplo:

#Digite um número: 7

# 7 x 1 = 7
# 7 x 2 = 14
# 7 x 3 = 21
...
# 7 x 10 = 70

n = int(input("Digite um número: "))
print("")
print("---Tabuada---")
#Processamento

for i in range(1 , 11):
  print(f"{n} x {i} = {n * i}")
```

```python
#QUESTÃO 04 – Fatorial (4 pontos)

#Solicite ao usuário um número inteiro positivo e calcule
#seu fatorial utilizando um laço de repetição.

#Exemplo:

#Digite um número: 5

#Fatorial = 120

#Pois:

# 5! = 5 × 4 × 3 × 2 × 1

#Observação:
#O programa deve validar se o número informado é positivo.

#entrada

n = int(input("Digite um número positivo: "))

while n < 0:
  print("Digite números positvos!")
  n = int(input("Digite um número positivo: "))

print("")

fatorial = 1

#processamento

for i in range(1, n + 1):
  fatorial *= i
print(f"O fatorial de {n} é {fatorial}")
```

```python
#QUESTÃO 05 – Caixa Eletrônico

#Um caixa eletrônico possui notas de:

# R$ 100
# R$ 50
# R$ 20
# R$ 10
# R$ 5
# R$ 2
# R$ 1

#Solicite ao usuário o valor do saque e informe a menor
#quantidade possível de notas para realizar o pagamento.

#Exemplo:

#Digite o valor do saque: 287

#Saída esperada:

# 2 nota(s) de R$100
# 1 nota(s) de R$50
# 1 nota(s) de R$20
# 1 nota(s) de R$10
# 1 nota(s) de R$5
# 1 nota(s) de R$2

#Requisitos:
# - Utilizar laços de repetição.
# - Não utilizar listas.
# - Exibir apenas as notas utilizadas.

#entrda

valor_saque = int(input("Digite o valor do saque: R$ "))
valor_restante = valor_saque

print(f"Para o saque de R$ {valor_saque}, as notas serão:")

dominacoes = (100, 50, 20, 10, 5, 2, 1)

for nota in dominacoes:
  if valor_restante >= nota:
    quantidade = valor_restante // nota
    if quantidade > 0:
      print(f"{quantidade} nota(s) de R$ {nota}")
    valor_restante %= nota

if valor_restante > 0:
  print("Não foi possível sacar o valor exato com as notas disponíveis.")
```
