```python
#operadores aritimeticos

#fazer um programa que o usuario digita 02 numeros.
#saida mostrar operações aritimeticas.


#entrada

a = int(input("digite um numero: "))
b = int(input("digite um numero: "))

#processamento

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
div_int = a //b
resto = a % b
potencia = a ** b

#saida

print("")
print("----------------------------calculadora------------------------")
print("")
print("--------------------------------soma---------------------------")
print("")
print(f"o resultado de {a} mais {b} é igual a: {soma}")
print("")
print("------------------------------subtracao------------------------")
print("")
print(f"o resultado de {a} menos {b} é igual a: {subtracao}")
print("")
print("----------------------------multiplicacao----------------------")
print("")
print(f"o resultado de {a} vezes {b} é igual a: {multiplicacao}")
print("")
print("-------------------------------divisao-------------------------")
print("")
print(f"o resultado de {a} dividido {b} é igual a: {divisao}")
print("")
print("----------------------------divisao inteira---------------------")
print("")
print(f"o resultado da divisao inteira de {a} dividido {b} é igual a: {div_int}")
print("")
print("--------------------------------resto---------------------------")
print("")
print(f"o resultado do resto de {a} dividido {b} é igual a: {resto}")
print("")
print("-------------------------------potencia-------------------------")
print("")
print(f"o resultado da potenciacao de {a} elevado a {b} é igual a: {potencia}")
```

```python
#Fazer um programa que converte horas em minutos e horas em segundos.
#entrada
hora = int(input("digite a hora: "))

#processamento

minutos = hora * 60
segundos = hora * 3600

#saida
print(f"{hora} horas é igual a {minutos} minutos ")
print(f"{hora} horas é igual a {segundos} segundos")
```

```python
#Fazer um programa que converte km em metros
#entrada
km = int(input("digite a distãncia: "))

#processamento
metros = km * 1000

#saida

print(f"{km} km é igual a {metros} metros")
```
