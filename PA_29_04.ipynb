```python
#Escrever um programa que o uuario digita uma palavra e o programa verifica se a palavra é uma palindroma e imprimi

#entrada
p = str(input("Digite uma palavra: "))
invertida = ""

for i in (p):
  invertida = i + invertida

if p == invertida:
  print(f"{p} é uma palavra palindroma.")

else:
  print(f"{p} não é uma palavra palindroma.")
```

```python
#Peça um número ao usuario e o programa irá separar números pares e impares
#exemplo de saida:

#pares: 0, 2, 4, 6 ...
#impares: 1, 3, 5 ...

#entrada

n = int(input("Digite um numero: "))

#processamento

print("pares")
for i in range(0, n, 2):
  print(i)

print("impares")
for i in range(1, n, 2):
  print(i)
```
