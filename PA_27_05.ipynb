```python
# jogo menu inteligente

# criar um sistema que funcione ate o usuário escolher sair

# menu

# 1 - par/impar
# 2 - positivo/negativo
# 3 - maior
# 4 - sair


opcao = 0

try:

  while opcao != 4:
    print("")
    print("---- MENU----")
    print("1 - par/impar")
    print("2 - positivo/negativo")
    print("3 - maior")
    print("4 - sair")
    print("")

    opcao = int(input("Digite uma opção: "))

    if opcao == 1:
      num = int(input("Digite um número: "))

      if num % 2 == 0:
        print(f" o número {num} é par")

      else:
        print(f" o número {num} é impar")


    elif opcao == 2:
      num = int(input("Digite um número: "))

      if num > 0:
        print(f" o número {num} é positivo")

      elif num == 0:
        print(f" o número é zero")

      else:
        print(f" o número {num} é negativo")


    elif opcao == 3:
      a = int(input("Digite o primeiro número: "))
      b = int(input("Digite o segundo número: "))
      if a > b:
        print(f"O número {a} é maior que o número {b}")

      elif a < b:
        print(f"O número {b} é maior que o número {a}")

      else:
        print(f" os números são iguais")


    else:
      print("O programa encerrou!")

except:
  print("Digite apenas números!")
```

```python
# urna eletronica

# criar um sistema de votação de três candidatos

# votar no candidato A
# votar no candidato B
# votar no candidato C
# votar branco
# sair

# o programa deve repitir ate encerrar a votação
# contar votos de cada candidato
# contar votos brancvos e nulos
# ao encerrar mostrar:

# total de votos
# total de votos por candidato
# total de votos brancos
# total de votos nulos
# vencedor da eleição

# regras extras:

# se houver empate entre os candidatos: mostrar eleição empatada
# se não hover votos válidos, mostrar: nenhum candidato recebeu voto



opcao = 0
votos_a = 0
votos_b = 0
votos_c = 0
votos_brancos = 0

while True:
    print("\n--- Urna Eletrônica ---")
    print("1 - Candidato A")
    print("2 - Candidato B")
    print("3 - Candidato C")
    print("4 - Voto Branco")
    print("5 - sair")
    print("")

    opcao = int(input("Digite uma opção: "))

    if opcao == '1':
            votos_a += 1
    elif opcao == '2':
            votos_b += 1
    elif opcao == '3':
            votos_c += 1
    elif opcao == '4':
            votos_brancos += 1
    elif opcao == 5:
      print("Votação encerrada!")
      break

 # calculos
total = votos_a + votos_b + votos_c + votos_brancos
votos_validos = votos_a + votos_b + votos_c
max_votos = max(votos_validos)

print("")
print("---resultado final---")
print(f"total de votos: {total}")
print(f"votos válidos: {votos_validos}")
print(f" votos no candidato A: {votos_a}")
print(f" votos no candidato B: {votos_b}")
print(f" votos no candidato C: {votos_c}")
print(f" votos em branco: {votos_brancos}")
print("")

    #regras extras

if votos_validos == 0:
      print(" Nenhum candidato rescebeu voto!")

else:
        # Verifica se há empate entre os que tiveram mais votos
        if votos_validos.count(max_votos) > 1:
            print("Eleição empatada")
        else:
            if votos_a == max_votos:
                print("Vencedor: Candidato A")
            elif votos_b == max_votos:
                print("Vencedor: Candidato B")
            else:
                print("Vencedor: Candidato C")
```
